```python
# Calculate Q1 and Q3
Q1 = Customer_dataset['MntTotal'].quantile(0.25)
Q3 = Customer_dataset['MntTotal'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers
outliers = Customer_dataset[( Customer_dataset['MntTotal'] < lower_bound) | ( Customer_dataset['MntTotal'] > upper_bound)]

# Print outliers
print(outliers)
```

```python
# Define lower and upper percentile limits
lower_cap = Customer_dataset['MntTotal'].quantile(0.01)
upper_cap = Customer_dataset['MntTotal'].quantile(0.99)

# Cap outliers
Customer_dataset['MntTotal'] = np.where(Customer_dataset['MntTotal'] < lower_cap, lower_cap, Customer_dataset['MntTotal'])
Customer_dataset['MntTotal'] = np.where(Customer_dataset['MntTotal'] > upper_cap, upper_cap, Customer_dataset['MntTotal'])
```

```python
def get_marital_status(row):
    if row['marital_Divorced'] == 1:
        return 'Divorced'
    elif row['marital_Married'] == 1:
        return 'Married'
    elif row['marital_Single'] == 1:
        return 'Single'
    elif row['marital_Together'] == 1:
        return 'Together'
    elif row['marital_Widow'] == 1:
        return 'Widow'
    else:
        return 'Unknown'
Customer_dataset['Marital Status'] = Customer_dataset.apply(get_marital_status, axis=1)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Marital Status', y='MntTotal', data= Customer_dataset, estimator='mean', palette='coolwarm')
plt.title('Average MntTotal Content by different Marital Status')
plt.show()
```

```python
def get_relationship(row):
    if row['marital_Married'] ==1:
        return 1
    elif row['marital_Together'] == 1:
        return 1
    else:
        return 0
Customer_dataset['In_relationship'] = Customer_dataset.apply(get_relationship, axis=1)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='In_relationship', y='Income', data= Customer_dataset, estimator='mean', palette='coolwarm')
plt.title('Average Income Content by different Marital Status')
plt.show()
```

```python
X = Customer_dataset[['Income','MntTotal','In_relationship']]
```

```python
X
```

```python
#Clustering
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

```python
#Taking 5 clusters
km1=KMeans(n_clusters=5)
#Fitting the input data
km1.fit(X)
#predicting the labels of the input data
y=km1.predict(X)
#adding the labels to a column named label
Customer_dataset["label"] = y
```
